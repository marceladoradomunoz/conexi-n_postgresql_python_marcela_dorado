# **DESARROLLO PRUEBA FINAL MÓDULO PYTHON**

```python
#PREGUNTA 1: GENERACIÓN DE FUNCIÓN LEER_TABLA Y CREACIÓN DE DF INDEPENDIENTES

def leer_tabla(tabla, engine):
  query = f'''
  SELECT *
    FROM {tabla}
  '''
  db_table = pd.read_sql(query, con=engine)
  print(db_table)
  return db_table

df_orders = leer_tabla("orders", engine)
df_orderdetails = leer_tabla('orderdetails', engine)
df_customers = leer_tabla('customers', engine)
df_products = leer_tabla('products', engine)
df_employeed = leer_tabla('employees', engine)
```

# **CREACIÓN DE FUNCIÓN LEER_TABLA**

####FUNCIÓN CUYO OBJETO ES LEER LAS TABLAS DEL ARCHIVO CLASSICMODELS.SLQ Y POSTERIORMENTE CONVERTIRLAS EN DATAFRAME


---



1.   Uso de palabra reservada def a objeto de definir la función denominándola
     leer_tabla, agregándo los parámetros solicitados (tabla, engine).
2.   El parámetro tabla indica la tabla a leer, mientras que el parámetro  
     engine hace referencia a la conexión o puente entre Python y PostgreSQL en entorno Google Colaboratory.
3.   Uso de la función query para realizar consulta en lenguaje SQL.
4.   Creación variable db_table y utilización de la función pd.read_sql con los
     parámetros query y engine, a objeto que transforme en una base de datos la lectura de la query utilizando funciones propias de Pandas mediante el puente de conexión creado con la función create_engine al cual previamente denominamos con.
5.   Uso de función print para imprimir en pantalla db_table.
6.   Uso de la función return a objeto que guarde en la base de datos db_table.
7.   Posteriormente se crean los DataFrame uno a uno con la función y sus
     respectivos parámetros.

```python
#PREGUNTA 2: CRUCE DE DATAFRAMES USANDO FUNCIÓN VALIDATE

df_u1 = pd.merge(df_orderdetails, df_orders, on = 'orderNumber', how = 'inner', validate = 'many_to_one')
df_u2 = pd.merge(df_u1, df_products, on = 'productCode', how = 'inner', validate = 'many_to_one')
df_u3 = pd.merge(df_u2, df_customers, on = 'customerNumber', how = 'right', validate = 'many_to_one' )
df_sales_report = pd.merge(df_u3, df_employeed, left_on = 'salesRepEmployeeNumber', right_on = 'employeeNumber', how = 'left', validate = 'many_to_one')
df_sales_report.head(1)

#Uso función merge (inner, right y left) a objeto de unir los DF resultantes en df_sales_report.
```

```python
#PREGUNTA 3: AGREGACIÓN DE COLUMNAS 'VENTA', 'COSTO' Y 'GANANCIA AL DF CREADO

df_sales_report['venta'] = df_sales_report.apply(lambda x: x['quantityOrdered'] * x['priceEach'], axis=1)
df_sales_report

df_sales_report['costo'] = df_sales_report.apply(lambda x: x['quantityOrdered'] * x['buyPrice'], axis=1)
df_sales_report.head(1)

df_sales_report['ganancia'] = df_sales_report.apply(lambda x: x['venta'] - x['costo'], axis=1)
df_sales_report.head(1)

#Creación de las columnas solicitadas utilizando función apply y expresiones lambda.
#Utilización parámetro axis = 1 a objeto que se agreguen en columnas en el df_sales_report.
```
